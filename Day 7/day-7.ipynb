```python
# Hangman Project

import random

stages = [
    r'''
     _______
     |/      |
     |      (_)
     |      \|/
     |       |
     |      /|\
     |
    ''',
    r'''
     _______
     |/      |
     |      (_)
     |      \|/
     |       |
     |      / 
     |
    ''',
    r'''
     _______
     |/      |
     |      (_)
     |      \|/
     |       |
     |      
     |
    ''',
    r'''
     _______
     |/      |
     |      (_)
     |      \|/
     |       
     |      
     |
    ''',
    r'''
     _______
     |/      |
     |      (_)
     |      \|
     |       
     |      
     |
    ''',
    r'''
     _______
     |/      |
     |      (_)
     |      
     |       
     |      
     |
    ''',
    r'''
     _______
     |/      
     |      
     |      
     |       
     |      
     |
    '''
]


logo = '''
 _                                             
| |                                            
| |__   __ _ _ __   __ _ _ __ ___   __ _ _ __  
| '_ \ / _` | '_ \ / _` | '_ ` _ \ / _` | '_ \ 
| | | | (_| | | | | (_| | | | | | | (_| | | | |
|_| |_|\__,_|_| |_|\__, |_| |_| |_|\__,_|_| |_|
                    __/ |                      
                   |___/  
'''

word_list = [
    "bulb", "messi", "camel", "python", "hangman", "guitar", "pencil",
    "computer", "keyboard", "monitor", "program", "algorithm", "notebook",
    "laptop", "elephant", "tiger", "mountain", "river", "ocean", "butterfly",
    "chocolate", "airplane", "bicycle", "coffee", "sunflower", "notebook",
    "umbrella", "rainbow", "diamond", "jungle", "library", "backpack", "window",
    "flower", "star", "planet", "soccer", "basketball", "football"
]


lives = 6
print(logo)

chosen_word = random.choice(word_list)
# print(chosen_word)

word_length = len(chosen_word)
placeholder = ""

for position in range(word_length):
    placeholder += "_"

print("Word to guess: " + placeholder)

game_over = False
correct_letters = []

while not game_over:
    print(f"****************************{lives}/6 LIVES LEFT****************************")
    guess = input("Guess a letter: ").lower()
    print(guess)

    if guess in correct_letters:
        print(f"You already guessed {guess}")

    display = ""

    for letter in chosen_word:
        if letter == guess:
            display += letter
            # print("You guessed right.")
            correct_letters.append(letter)
        elif letter in correct_letters:
            display += letter

        else:
            display += "_"

    print("Word to guess: " + display)

    if guess not in chosen_word:
        lives -= 1
        print(f"You guessed {guess}, that's not in the word. You lose a life.")

        if lives == 0:
            game_over = True
            print(f"****************************IT WAS {chosen_word}! YOU LOSE****************************")

    print(stages[lives])

    if "_" not in display:
        game_over = True
        print("****************************YOU WIN****************************")
```
