```python
# List = [] ordered and changeable. Duplicates OK
# Set = {} unordered and immutable, but Add/ Remove OK. NO duplicates
# Tuple = () ordered and unchangeable, Duplicates OK. FASTER
```

```python

fruits = ["Apple", "Orange", "Banana", "Coconut"]
print(fruits)
print(fruits[0])
print(fruits[0:3])
print(fruits[::2])
print(fruits[::-1])

for fruit in fruits:
    print(fruit)

print(dir(fruits))
print(help(fruits))
```

```python
# List

fruits = ["Apple", "Orange", "Banana", "Coconut"]
print(fruits)
print(len(fruits))
print("Apple" in fruits)
print("Pineapple" in fruits)

fruits[0] = "Pineapple"
print(fruits)

fruits.append("Mango")
print(fruits)

fruits.remove("Orange")
print(fruits)

fruits.insert(0, "Apple")
print(fruits)

fruits.sort()
print(fruits)

fruits.reverse()
print(fruits)

print(fruits.index("Apple"))
print(fruits.index("Mango"))
# print(fruits.index("Peach"))

print(fruits.count("Apple"))
print(fruits.count("Peach"))

fruits.clear()
print(fruits)
```

```python
# List

fruits = {"Apple", "Orange", "Banana", "Coconut"}

print(fruits)
print(dir(fruits))
print(help(fruits))
```

```python
# List

fruits = {"Apple", "Orange", "Banana", "Coconut", "Coconut"}

print(fruits)
print(len(fruits))
print("Pineapple" in fruits)

# print(fruits[0]) # Set is unordered 
fruits.add("Pineapple")
print(fruits)

fruits.remove("Apple")
print(fruits)

fruits.pop() # Randomly remove element
print(fruits)


fruits.clear()
print(fruits)
```
